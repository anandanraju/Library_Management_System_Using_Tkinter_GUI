### Publisher Table

```python
from tkinter import *
import tkinter.ttk as ttk
import tkinter.messagebox as tkMessageBox
import sqlite3
```

```python
def Database():
    global file, cursor
    file = sqlite3.connect('D:\\DBMS\\Library-Management-System.sqlite')
    cursor = file.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS PUBLISHER_TABLE (PUBLISHER_NAME TEXT PRIMARY KEY AUTOINCREMENT NOT NULL,\
                    PUBLISHER_ADDRESS VARCHAR,PUBLISHER_PHONENUMBER VARCHAR)")

def DisplayForm():
    display_screen = Tk() # Create Window
    display_screen.geometry('1400x700') #Setting Width & Height for Window
    display_screen.title('Library Management System')

    global tree
    global SEARCH
    global publisher_name,publisher_address,publisher_phonenumber

    SEARCH = StringVar()
    publisher_name = StringVar()
    publisher_address = StringVar()
    publisher_phonenumber = StringVar()

    #Creating Frame for Layout
    #TopView frame for Heading
    TopViewForm = Frame(display_screen,width=600,bd=1,relief=SOLID)
    TopViewForm.pack(side=TOP,fill=X)
    #First Left Frame for Registration Form
    LForm = Frame(display_screen,width=500,bg='#e3e3e3')
    LForm.pack(side=LEFT,fill=Y)
    #Second Left Frame for Search Form
    LeftViewForm = Frame(display_screen,width=300,bg='#b3b3b3',highlightbackground='black',highlightthickness=2)
    LeftViewForm.pack(side=LEFT,fill=Y)
    #Third Left Frame for Search Form
    LeftViewForm1 = Frame(display_screen,width=300,highlightbackground='black')
    LeftViewForm1.pack(side=LEFT,fill=Y)
    # MidFrame for Displaying Library Records
    MidViewForm = Frame(display_screen,width=1000,highlightbackground='black',highlightthickness=1)
    MidViewForm.pack(side=RIGHT)

    #Label for Heading
    lbl_text = Label(TopViewForm,text='LIBRARY MANAGEMENT SYSTEM',font=('Sans-Serif',15),width=120,bg="lightblue",fg="black")
    lbl_text.pack(fill=X)

    #Creating Registration Form in First Left Frame
    Label(LForm,text="PUBLISHER NAME ",font=("Arial",8)).pack(side=TOP)
    Entry(LForm,font=('Arial',10,'bold'),textvariable=publisher_name).pack(side=TOP,padx=10,fill=X)
    Label(LForm,text="PUBLISHER ADDRESS ",font=('Arial',8)).pack(side=TOP)
    Entry(LForm,font=('Arial',10,'bold'),textvariable=publisher_address).pack(side=TOP,padx=10,fill=X)
    Label(LForm,text="PUBLISHER PHONE NUMBER ",font=('Arial',8)).pack(side=TOP)
    Entry(LForm,font=('Arial',10,'bold'),textvariable=publisher_phonenumber).pack(side=TOP,padx=10,fill=X)
    Button(LForm,text='SUBMIT',font=('Arial',10,'bold'),command=Register).pack(side=TOP,padx=10,pady=5,fill=X)

    #Creating Search Label and Entry in Third Frame
    lbl_txtsearch = Label(LeftViewForm,text='ENTER PUBLISHER NAME \n TO SEARCH',font=('Verdana',8),bg='#b3b3b3')
    lbl_txtsearch.pack()

    #Creating Search Entry
    search = Entry(LeftViewForm,textvariable=SEARCH,font=('Verdana',15),width=10)
    search.pack(side=TOP,padx=10,fill=X)

    #Creating Search Entry
    btn_search = Button(LeftViewForm,text='SEARCH',command=SearchRecord)
    btn_search.pack(side=TOP,padx=10,pady=10,fill=X)

    #Creating View Button
    btn_view = Button(LeftViewForm,text='VIEW ALL',command=DisplayData)
    btn_view.pack(side=TOP,padx=10,pady=10,fill=X)

    #Creating Reset Button
    btn_reset = Button(LeftViewForm,text='RESET',command=Reset)
    btn_reset.pack(side=TOP,padx=10,pady=10,fill=X)
    
    #Creating Search Label and Entry in Third Frame
    lbl_txtsearch1 = Label(LeftViewForm1,text='ENTER PUBLISHER PHONE NUMBER \n TO SEARCH',font=('Verdana',8))
    lbl_txtsearch1.pack()

    #Creating Search Entry
    search1 = Entry(LeftViewForm1,textvariable=SEARCH,font=('Verdana',15),width=10)
    search1.pack(side=TOP,padx=10,fill=X)

    #Creating Search Entry
    btn_search1 = Button(LeftViewForm1,text='SEARCH',command=SearchRecord)
    btn_search1.pack(side=TOP,padx=10,pady=10,fill=X)

    #Creating View Button
    btn_view1 = Button(LeftViewForm1,text='VIEW ALL',command=DisplayData)
    btn_view1.pack(side=TOP,padx=10,pady=10,fill=X)

    #Creating Reset Button
    btn_reset1 = Button(LeftViewForm1,text='RESET',command=Reset)
    btn_reset1.pack(side=TOP,padx=10,pady=10,fill=X)

    
    bk_tbl_btn = Button(LForm,text='BOOK TABLE',font=('Bold',10),fg='black',bd=2,bg='#b3b3b3')
    bk_tbl_btn.pack(side=BOTTOM,padx=10,pady=20,fill=X)
    
    pubr_tbl_btn = Button(LForm,text='PUBLISHER TABLE',font=('Bold',10),fg='black',bd=2,bg='#b3b3b3')
    pubr_tbl_btn.pack(side=BOTTOM,padx=10,pady=10,fill=X)
    
    share_btn = Button(LeftViewForm1,text='SHARE',font=('Bold',10),fg='black',bd=2,bg='#b3b3b3')
    share_btn.pack(side=BOTTOM,padx=10,pady=10,fill=X)
    
    opt_btn = Button(LeftViewForm1,text='OPTIONS',font=('Bold',10),fg='black',bd=2,bg='#b3b3b3')
    opt_btn.pack(side=BOTTOM,padx=10,pady=10,fill=X) 
    
    menu_btn = Button(LeftViewForm1,text='MENU',font=('Bold',10),fg='black',bd=2,bg='#b3b3b3')
    menu_btn.pack(side=BOTTOM,padx=10,pady=10,fill=X)
    
    nxt_btn = Button(MidViewForm,text='NEXT PAGE',font=('Bold',10),fg='black',bd=3,bg='#b3b3b3')
    nxt_btn.pack(side=BOTTOM,padx=10,pady=10,fill=X)
    
   

    #Setting Scrollbar
    scrollbarx = Scrollbar(LeftViewForm,orient=HORIZONTAL)
    scrollbary = Scrollbar(MidViewForm,orient=VERTICAL)
    tree= ttk.Treeview(MidViewForm,columns=('PUBLISHER_NAME','PUBLISHER_ADDRESS','PUBLISHER_PHONENUMBER'),
                      selectmode='extended',height=70,yscrollcommand=scrollbary.set,xscrollcommand=scrollbarx.set)
    scrollbary.config(command=tree.yview)
    scrollbary.pack(side=RIGHT,fill=Y)
    scrollbarx.config(command=tree.xview)
    scrollbarx.pack(side=BOTTOM,fill=X)

    #Setting Headings for the Columns
    tree.heading('PUBLISHER_NAME', text="Publisher Name", anchor=W)
    tree.heading('PUBLISHER_ADDRESS', text="Publisher Address")
    tree.heading('PUBLISHER_PHONENUMBER', text="Publisher Phone Number", anchor=W)
    #setting width of the columns
    tree.column('#0', stretch=NO, minwidth=0, width=0)
    tree.column('#1', stretch=NO, minwidth=0, width=180)
    tree.column('#2', stretch=NO, minwidth=0, width=500)
    tree.pack()
    DisplayData()

def Register():
    Database() #Getting Form Data
    p_name1=publisher_name.get()
    p_add1=publisher_address.get()
    p_phno1=publisher_phonenumber.get()
    if p_name1=='' or p_add1=='' or p_phno1=='':
        tkMessageBox.showinfo("Warning","Fill the Empty Field")
    else:
        file.execute('INSERT INTO PUBLISHER_TABLE(PUBLISHER_NAME,PUBLISHER_ADDRESS,PUBLISHER_PHONENUMBER) VALUES (?,?,?)',(p_name1,p_add1,p_phno1));
        file.commit()

        tkMessageBox.showinfo("Message","Stored Successfully")
        DisplayData()
        file.close()

def Reset():
    tree.delete(*tree.get_children()) #Clear Current Data from Table
    DisplayData() 
    SEARCH.set("")
    publisher_name.set("")
    publisher_address.set("")
    publisher_phonenumber.set("")

def Delete():
    DataBase()
    if not tree.selection():
        tkMessageBox.showwarning("Warning","Select Data to Delete")
    else:
        result = tkMessageBox.askquestion("Confirm","Are You Sure to Delete this Record ?",icon='Warning')
        if result=="yes":
            curItem = tree.focus()
            contents = (tree.item(curItem))
            selecteditem = contents['values']
            tree.delete(curItem)
            cursor = file.execute("DELETE FROM PUBLISHER_TABLE WHERE PUBLISHER_NAME = %s"%selecteditem[0])
            file.commit()
            file.close()
            cursor.close()

def SearchRecord():
    Database()
    if SEARCH.get() != "":
        tree.delete(*tree.get_children())
        cursor=file.execute("SELECT * FROM PUBLISHER_TABLE WHERE PUBLISHER_NAME LIKE ?",('%' + str(SEARCH.get()) + '%',))
        fetch = cursor.fetchall()
        for data in fetch:
            tree.insert('','end',values=(data))
        cursor.close()
        file.close()


def DisplayData():
    Database()
    tree.delete(*tree.get_children())
    cursor=file.execute("SELECT * FROM PUBLISHER_TABLE")
    fetch = cursor.fetchall()
    for data in fetch:
        tree.insert('','end',values=(data))
    cursor.close()
    file.close()

DisplayForm()
if __name__=='__main__':
    mainloop()
```
